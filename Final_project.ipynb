## [Sourcetree](https://www.sourcetreeapp.com/)

Sourcetree is a free Git and Mercurial desktop client, developed by Atlassian, which provides a graphical user interface (GUI) for managing and interacting with version control repositories. 

With Sourcetree, users can easily clone, create, and manage Git and Mercurial repositories. It simplifies many complex version control tasks by providing an intuitive and user-friendly interface. Some of its key features include:

1. Repository Management: Sourcetree allows users to clone, create, and manage repositories from various hosting services like GitHub, Bitbucket, and GitLab. It provides an easy way to browse repositories, view branches, and access commit history.

2. Visual Branching and Merging: Sourcetree offers a visual representation of branches, making it easier to understand and manage complex branching structures. Users can visually compare branches, merge changes, and resolve conflicts within the GUI.

To download sourtree, click on the "Sourcetree" title. After that, link your git account (from gitlab, github, bitbucket, or others) to the software and once it is done, you will be ready to start!
